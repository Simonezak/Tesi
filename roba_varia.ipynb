```python
import random
import networkx as nx
from wntr.network import WaterNetworkModel
from wntr.network.io import write_inpfile


# ============================================================
# CONFIG
# ============================================================

inp_path = r"C:\Users\nephr\Desktop\Uni-Nuova\Tesi\Networks-found-FINAL\20x20_branched_BSD_4_2.inp"  # <-- METTI QUI IL PATH GIUSTO
remove_prob = 0.05


# ============================================================
# RANDOM BRANCHED NETWORK GENERATION
# ============================================================

while True:
    # Carica ogni volta la rete originale
    wn = WaterNetworkModel(inp_path)

    # Lista delle pipe (copiata per sicurezza)
    edges_list = list(wn.pipe_name_list)

    # Rimozione casuale delle pipe
    for edge in edges_list:
        if random.random() < remove_prob:
            wn.remove_link(edge)

    # Controllo di connettivit
    G = wn.get_graph().to_undirected()

    if nx.is_connected(G):
        break
    else:
        print("Generated disconnected network, retrying...")


# ============================================================
# VISUALIZATION & USER CONFIRMATION
# ============================================================

#wn.plot_network()

# ---------------- pipes ----------------
pipe_names = wn.pipe_name_list
num_pipes = len(pipe_names)

# ---------------- graph ----------------
G = wn.get_graph().to_undirected()

# Branch pipes = edges whose removal disconnects the graph
bridges = list(nx.bridges(G))
num_branch_pipes = len(bridges)

ratio = num_branch_pipes / num_pipes if num_pipes > 0 else 0.0

c = input("Ci piace il network? (y/n) ")

if c.lower() == "y":
    write_inpfile(wn, r"C:\Users\nephr\Desktop\Uni-Nuova\Tesi\Networks-found-FINAL\20x20_branched_BSD_4_3.inp")
    print("Network saved as 10x10_branched.inp")
```

```python
import wntr
import networkx as nx

def print_network_stats(inp_path):
    """
    Print basic structural statistics of a water distribution network.
    """

    # ---------------- load network ----------------
    wn = wntr.network.WaterNetworkModel(inp_path)

    # ---------------- nodes ----------------
    num_junctions = len(wn.junction_name_list)
    num_reservoirs = len(wn.reservoir_name_list)
    num_tanks = len(wn.tank_name_list)
    num_nodes = num_junctions + num_reservoirs + num_tanks

    # ---------------- pipes ----------------
    pipe_names = wn.pipe_name_list
    num_pipes = len(pipe_names)

    # ---------------- graph ----------------
    G = wn.get_graph().to_undirected()

    # Branch pipes = edges whose removal disconnects the graph
    bridges = list(nx.bridges(G))
    num_branch_pipes = len(bridges)

    ratio = num_branch_pipes / num_pipes if num_pipes > 0 else 0.0

    # ---------------- print ----------------
    print("\n================= NETWORK STATISTICS =================")
    print(f"Total nodes              : {num_nodes}")
    print(f"  - Junctions            : {num_junctions}")
    print(f"  - Reservoirs           : {num_reservoirs}")
    print(f"  - Tanks                : {num_tanks}")

    print(f"\nTotal pipes              : {num_pipes}")
    print(f"Branch pipes             : {num_branch_pipes}")
    print(f"Ratio (Branch / Total)   : {ratio:.3f}")


# ============================================================
# MAIN
# ============================================================

if __name__ == "__main__":

    inp_path = r"C:\Users\nephr\Desktop\Uni-Nuova\Tesi\Networks-found-FINAL\modena_BSD.inp"
    print_network_stats(inp_path)
```
